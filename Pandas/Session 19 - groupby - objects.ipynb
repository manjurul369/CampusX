```python
# find ranking of each movie in the group according to IMDB score

def rank_movie(group):
  group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
  return group
```

```python
genres.apply(rank_movie)
```

```python
# find normalized IMDB rating group wise
def normalize(group):
  group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
  return group
```

```python
genres.apply(normalize).loc['Action']
```

```python
# find the top 10 batsman in terms of runs
batter = ipl.groupby('batsman')
batter['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max no of six
ipl[ipl['batsman_runs'] == 6].groupby('batsman')['batsman_runs'].value_counts().sort_values(ascending=False).head(1)
```

```python
# find batsman with most number of 4's and 6's in last 5 overs
last_overs = ipl[ipl['over'] > 15]
mask = last_overs[(last_overs['batsman_runs'] == 4) | (last_overs['batsman_runs'] == 6)]
mask.groupby('batsman')['batsman_runs'].value_counts().sort_values(ascending=False).head(1)
```

```python
# find V Kohli's record against all teams
temp = ipl[ipl['batsman'] == 'V Kohli']
temp.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
# create a function that can return the highest score of any batsman
def highest(batsman):
  temp_df = ipl[ipl['batsman'] == batsman]
  return temp_df.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
```

```python
highest('CH Gayle')
```
