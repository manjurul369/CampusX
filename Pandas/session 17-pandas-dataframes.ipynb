```python
#how many super over finishes have occured
ipl[ipl['SuperOver'] == 'Y'].shape[0]
```

```python
#how many matches has CSK won in Kolkata
kolkata_matches = ipl[ipl['City'] == 'Kolkata']
kolkata_matches[kolkata_matches['WinningTeam'] == 'Chennai Super Kings']

ipl[(ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')] #using one line
```

```python
#toss winner is match winner percentage
ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]/ipl.shape[0] * 100
```

```python
movies[(movies['imdb_votes']>10000) & (movies['imdb_rating'] > 8)].shape[0]
```

```python
#Action movies with rating higher than 7.5
mask1 = movies['genres'].apply(lambda x: x.split('|')).apply(lambda x: 'Action' in x)

mask2 = movies['imdb_rating'] > 7.5

movies[mask1 & mask2].shape[0]
```

```python
#write a function that can return the tract record of 2 teams against each other
team1 = input("Team1 = ")
team2 = input("Team2 = ")

mask1 = (ipl['Team1'] == team1) & (ipl['Team2'] == team2)
mask10 = (ipl['Team1'] == team2) & (ipl['Team2'] == team1)
mask2 = ipl['WinningTeam'] == team1
mask3 = ipl['WinningTeam'] == team2

win1 = ipl[(mask1 | mask10) & mask2].shape[0]
print(f"{team1} won {win1} matches against {team2}")

win2 = ipl[(mask1 | mask10) & mask3].shape[0]
print(f"{team2} won {win2} matches against {team1}")
```
