### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
# scatter plot --> axes level function
sns.scatterplot(data = tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
```

```python
# scatter plot --> relplot --> square shape
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex', style='time', size='size')
```

```python
# line plot
gap = px.data.gapminder()
gap
```

```python
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
# using axes plot
sns.lineplot(data=temp_df, x = 'year', y='lifeExp')
```

```python
# using relplot
sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line')
```

```python
# hue --> style
temp_df = gap[gap['country'].isin(['India', 'Brazil', 'Germany', 'Pakistan'])]
temp_df
```

```python
sns.relplot(data=temp_df, x='year', y='lifeExp', hue='country', kind='line', style='continent', size='continent')
```

```python
sns.lineplot(data=temp_df, x='year', y='lifeExp', hue='country')
```

```python
# facet plot --> figure level function --> work with relplot
# it won't work with scatterplot and lineplot
sns.relplot(data=tips, x='total_bill', y='tip', kind='line', row='day', col='sex')
```

```python
# col wrap
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=4)
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
# figure level --> displot
# axes level --> histplot/kdeplot/rugplot
```

```python
# plotting univariant histogram
sns.histplot(data=tips, x='total_bill')
```

```python
sns.displot(kind='hist', data=tips, x='total_bill')
```

```python
# bin parameter
sns.displot(kind='hist', data=tips, x='total_bill', bins=20)
```

```python
# It's also possible to visualize the distribution of a categorical variable using the logic of a histogram
# Discrete bins are automatically set for categorical variable

# It's one kind of countplot
sns.displot(kind='hist', data=tips, x='day')
```

```python
# hue parameter
sns.displot(kind='hist', data=tips, x='total_bill', bins=20, hue='sex')
```

```python
# element --> Step
sns.displot(kind='hist', data=tips, x='total_bill', bins=20, hue='sex', element='step')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(kind='hist', data=titanic, x='age', hue='survived')
```

```python
# faceting using col and row --> not work on histplot function
sns.displot(kind='hist', data=titanic, x='age', col='sex')
```

```python
# kdeplot
# rather than using discrete bins, a KDE plot smooths the observations with a Guassian kernal, producing a continuous density estimate
sns.kdeplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill', kind='kde')
```

```python
# hue --> fill
plt.figure(figsize=(15,8)) # size changes for axes level plots
sns.kdeplot(data=tips, x='total_bill', hue='sex', fill=True)
```

```python
# hue --> fill
sns.displot(data=tips, x='total_bill', hue='sex', fill=True, kind='kde', height=10, aspect=2) # size changes for figure level plots --> height, aspects
```

```python
# rugplot
# plot marginal distributions by drawing ticks along the x and y axis
# This function is intended to complement others plot by showing the location of individual observations in an unobtrusive way
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
```

```python
# Bivariate Histogram
# A bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observations within each rectangle with the fill color

sns.histplot(data=tips, x='total_bill', y='tip')
```

```python
# Bivariate kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Guassian

sns.kdeplot(data=tips, x='total_bill', y='tip')
```

### 2. Matrix Plot

- Heatmap
- Clustermap

```python
# Heatmap

# Plot rectangular data as a color-encoded matrix

temp_df = gap.pivot(index='country', columns='year', values='lifeExp')

plt.figure(figsize=(15,9))
sns.heatmap(data=temp_df)
```

```python
# annot
plt.figure(figsize=(15,9))
sns.heatmap(gap[gap['continent'] == 'Europe'].pivot(index='country', columns='year', values='lifeExp'), annot=True)
```

```python
# linewidth
plt.figure(figsize=(15,9))
sns.heatmap(gap[gap['continent'] == 'Europe'].pivot(index='country', columns='year', values='lifeExp'), annot=True, linewidths=0.6)
```

```python
# cmap
plt.figure(figsize=(15,9))
sns.heatmap(gap[gap['continent'] == 'Europe'].pivot(index='country', columns='year', values='lifeExp'), annot=True, cmap='viridis')
```

```python
# Clustermap

# Plot a matrix dataset as a hierarchiacally-clustered heatmap

# This function requres scipy to be availab

iris = px.data.iris()
sns.clustermap(iris.iloc[:,:4])
```
