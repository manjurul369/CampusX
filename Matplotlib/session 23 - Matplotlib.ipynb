### 2D Line plot

```python
# plotting a simple function
price = [48000, 54000, 57000, 49000, 49000, 45000, 47000]
year = [2014, 2015, 2016, 2017, 2018, 2019, 2020]

plt.plot(year, price)
```

```python
# from a pandas dataframe
batsman = pd.read_csv('/content/drive/MyDrive/Datasets/sharma-kohli.csv')
batsman

plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# multiple line
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
# labels and titles
plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])

plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors (hex) and line (width and style) and marker (size)
plt.plot(batsman['index'], batsman['V Kohli'], color='green')
plt.plot(batsman['index'], batsman['RG Sharma'], color='#000')

plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# line styles and widths
plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='dashed', linewidth=3)
plt.plot(batsman['index'], batsman['RG Sharma'], color='#000', linestyle='dashdot', linewidth=2)

plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# markers
# line styles and widths
plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='dashed', linewidth=3, marker='D') # D --> Diamond
plt.plot(batsman['index'], batsman['RG Sharma'], color='#000', linestyle='dashdot', linewidth=2, marker='.', markersize=20)

plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend --> location
plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='dashed', linewidth=3, marker='D', label='Virat') # D --> Diamond
plt.plot(batsman['index'], batsman['RG Sharma'], color='#000', linestyle='dashdot', linewidth=2, marker='.', markersize=20, label='Rohit')

plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend(loc='upper center') # default --> best
```

```python
# limiting axes
price = [48000, 54000, 57000, 49000, 49000, 45000, 4700000]
year = [2014, 2015, 2016, 2017, 2018, 2019, 2020]

plt.plot(year, price)
plt.ylim(0, 75000)
plt.xlim(2015, 2017)
```

```python
# grid
plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='dashed', linewidth=3, marker='D', label='Virat') # D --> Diamond
plt.plot(batsman['index'], batsman['RG Sharma'], color='#000', linestyle='dashdot', linewidth=2, marker='.', markersize=20, label='Rohit')

plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend(loc='upper center')
plt.grid()
```

```python
# show
plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='dashed', linewidth=3, marker='D', label='Virat') # D --> Diamond
plt.plot(batsman['index'], batsman['RG Sharma'], color='#000', linestyle='dashdot', linewidth=2, marker='.', markersize=20, label='Rohit')

plt.title('Rohit Sharma VS Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend(loc='upper center')
plt.grid()
plt.show()
```

### Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart
children = [10,20,23,40,15]
colors = ['red', 'blue', 'green', 'yellow', 'pink']
plt.bar(colors, children, color='#000')
plt.title('Children Vs Their Favourite Color')
```

```python
# horizontal bar
plt.barh(colors, children, color='#000')
```

```python
# colors and labels
bat_season = pd.read_csv('/content/drive/MyDrive/Datasets/batsman_season_record.csv')
bat_season
```

```python
# multiple bar plots
plt.bar(bat_season['batsman'], bat_season['2015'])
plt.bar(bat_season['batsman'], bat_season['2016'])
plt.bar(bat_season['batsman'], bat_season['2017'])
```

```python
# multiple bar plots
plt.bar(bat_season['batsman'], bat_season['2015'], width=.3)
plt.bar(bat_season['batsman'], bat_season['2016'], width=.3)
plt.bar(bat_season['batsman'], bat_season['2017'], width=.4)
```

```python
plt.bar(np.arange(bat_season.shape[0]) - 0.2, bat_season['2015'], width=0.2, color='green')
plt.bar(np.arange(bat_season.shape[0]), bat_season['2016'], width=0.2, color='#000')
plt.bar(np.arange(bat_season.shape[0]) + 0.2, bat_season['2017'], width=0.2, color='red')

plt.xticks(np.arange(bat_season.shape[0]), bat_season['batsman'])
plt.show()
```

```python
# xticks --> replace x value names
```

```python
# a problem
children = [10,20,23,40,15]
colors = ['red red red red red red', 'blue blue blue blue blue', 'green green green green green', 'yellow', 'pink']
plt.bar(colors, children, color='#000')
plt.title('Children Vs Their Favourite Color')
```

```python
children = [10,20,23,40,15]
colors = ['red red red red red red', 'blue blue blue blue blue', 'green green green green green', 'yellow', 'pink']
plt.bar(colors, children, color='#000')
plt.title('Children Vs Their Favourite Color')
plt.xticks(rotation=70)
```

```python
# stacked bar chart
plt.bar(bat_season['batsman'], bat_season['2015'], bottom=(bat_season['2016'] + bat_season['2017']), label='2015')
plt.bar(bat_season['batsman'], bat_season['2016'], bottom=bat_season['2017'], label='2016')
plt.bar(bat_season['batsman'], bat_season['2017'], label='2017')

plt.legend()
plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# simple data
data = [23, 45, 100, 20, 25]
subjects = ['English', 'Science', 'Maths', 'History', 'Bangla']
plt.pie(data, labels=subjects)
plt.show()
```

```python
# dataset
gayle = pd.read_csv('/content/drive/MyDrive/Datasets/gayle-175.csv')
gayle
```

```python
plt.pie(gayle['batsman_runs'], labels=gayle['batsman'])
plt.show()
```

```python
# percentage
plt.pie(gayle['batsman_runs'], labels=gayle['batsman'], autopct='%0.1f%%')
plt.show()
```

```python
# colors
plt.pie(gayle['batsman_runs'], labels=gayle['batsman'], autopct='%0.1f%%', colors=['blue', 'green', 'red', 'yellow', 'black', 'orange'])
plt.show()
```

```python
# label color
plt.pie(gayle['batsman_runs'], labels=gayle['batsman'], autopct='%0.1f%%', colors=['blue', 'green', 'red', 'yellow', 'black', 'orange'], textprops={'color': 'red'})
plt.show()
```

```python
# explode shadow
plt.pie(gayle['batsman_runs'], labels=gayle['batsman'], autopct='%0.1f%%', explode=[0.1,0,0,0,0.2,0], shadow=True)
plt.show()
```
