## Annotations

```python
batters = pd.read_csv('/content/batter.csv')
batters
```

```python
sample_df = batters.head(100).sample(25, random_state=5)
sample_df
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.scatter(x,y)
plt.text(1,5,'Point One')
plt.text(2,6,'Point Two')
plt.text(3,7,'Point Three')
plt.text(4,8,'Point Four', fontdict={'size': 12, 'color': 'green'})
plt.show()
```

## Horizontal and Vertical Lines

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])

plt.axhline(130, color='green')
plt.axhline(140, color='red')
plt.axvline(30, color='green')

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

## Subplots

```python
# A different way to plot graphs
batters.head()
```

```python
plt.figure(figsize=(15,8))
plt.scatter(batters['avg'], batters['strike_rate'])
plt.title('Something')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')
```

```python
fig, ax = plt.subplots(figsize=(15,8))

ax.scatter(batters['avg'], batters['strike_rate'], color='red', marker='+')
ax.set_title('Something')
ax.set_xlabel('Avg')
ax.set_ylabel('Strike Rate')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(15,10))

ax[0].scatter(batters['avg'], batters['strike_rate'])
ax[0].set_title('Average Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].scatter(batters['avg'], batters['runs'])
ax[1].set_title('Average Vs Runs')
ax[1].set_xlabel('Avg')
ax[1].set_ylabel('Runs')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15,10))

ax[0,0].scatter(batters['avg'], batters['strike_rate'])
ax[0,1].scatter(batters['avg'], batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])
```

```python
fig = plt.figure(figsize=(20,8))

ax1 = fig.add_subplot(1,2,1)
ax1.scatter(batters['avg'], batters['strike_rate'])
ax1.set_title('Average Vs Strike Rate')
ax1.set_ylabel('Strike Rate')

ax2 = fig.add_subplot(1,2,2)
ax2.hist(batters['avg'])
ax2.set_title('Average Vs Strike Rate')
ax2.set_ylabel('Strike Rate')
```

## 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
yy.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(18,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(p, ax=ax)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(18,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(p, ax=ax)
```

```python
z = np.sin(xx) + np.log(yy)
fig = plt.figure(figsize=(18,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(p, ax=ax)
plt.show()
```

## Contour Plot

```python
z = xx**2 + yy**2
fig = plt.figure(figsize=(18,8))

ax = plt.subplot()

p = ax.contour(xx,yy,z, cmap='viridis')
fig.colorbar(p, ax=ax)
plt.show()
```

```python
z = xx**2 + yy**2
fig = plt.figure(figsize=(18,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z, cmap='viridis')
fig.colorbar(p, ax=ax)
plt.show()
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(18,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z, cmap='viridis')
fig.colorbar(p, ax=ax)
plt.show()
```

## Heatmap

```python
delivery = pd.read_csv('/content/IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run'] == 6)]
```

```python
grid = temp_df.pivot_table(index='overs', columns='ballnumber', values='batsman_run', aggfunc='count')
```

```python
plt.figure(figsize=(20,10))
plt.imshow(grid)
plt.xticks(np.arange(0,6), list(range(1,7)))
plt.yticks(delivery['overs'].unique(), list(range(1,21)))
plt.colorbar()
```
